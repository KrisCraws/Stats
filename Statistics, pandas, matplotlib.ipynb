```python
import pandas as pd
import matplotlib.pyplot as plt

wnba = pd.read_csv("WNBA Stats.csv")

# take 100 samples of 10 values each from our WNBA data set, and for each sample, measure the average points scored by a 
# player during the 2016-2017 season. Make it reproducible by varying the random_state using values btwn 0 and 99

working_list = []
for i in range(0,100):
    rand_samp = wnba['PTS'].sample(10, random_state = i)
    working_list.append(rand_samp.mean())

#calculates the population mean for pts scored by players
parameter = wnba['PTS'].mean()   

# display all the 100 sample means using a scatter plot. For the x-axis, use integers from 1 to 100 to designate the sample number. 
# Use the y-axis for the sample means
plt.scatter(range(1,101), working_list) 

plt.axhline(parameter) # draw a horizontal line that represents the average number of points for the population.
plt.show() # displays plot
```
